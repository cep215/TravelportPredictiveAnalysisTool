```python
import os
import pandas as pd, numpy as np
pd.set_option('display.max_colwidth', -1)
pd.set_option('display.max_rows', 100)
pd.set_option('display.max_columns', 200)

from pandas import DataFrame
from io import StringIO
import time, json
from datetime import date
import statsmodels.api as sm
from statsmodels.tsa.stattools import adfuller, acf, pacf
from statsmodels.tsa.arima_model import ARIMA
from statsmodels.tsa.seasonal import seasonal_decompose
from sklearn.metrics import mean_squared_error
import matplotlib.pylab as plt
%matplotlib inline
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 15, 6
```

# Read Data

```python
DATA_PATH = 'Desktop/Courses/Third\ Year/Group\ Project\ -\ Travelport/jupyter/'

def read_data(data_path = DATA_PATH):
    data_path = os.path.join(data_path, 'part-v001-o000-r-00000')
    df = pd.read_csv(data_path, names = ['Date', 'Nr_Of_Bookings'])
    return df
```

```python
bookings_df = read_data()
bookings_df.head()
```

## Create Time Series

```python
bookings_df['Date'] = pd.to_datetime(bookings_df['Date'])
date_indexed_bookings_df = bookings_df.set_index('Date')
ts = date_indexed_bookings_df['Nr_Of_Bookings']
ts = ts[1:-1] # first and last week may not be complete
```

```python
ts.head()
```

```python
ts.plot()
```

## Check Stationarity of Time Series

```python
def test_stationarity(timeseries):
    
    #Determing rolling statistics
    rolmean = pd.rolling_mean(timeseries, window=52)
    rolstd = pd.rolling_std(timeseries, window=52)

    #Plot rolling statistics:
    orig = plt.plot(timeseries, color='blue',label='Original')
    mean = plt.plot(rolmean, color='red', label='Rolling Mean')
    std = plt.plot(rolstd, color='black', label = 'Rolling Std')
    plt.legend(loc='best')
    plt.title('Rolling Mean & Standard Deviation')
    plt.show(block=False)
    
    #Perform Dickey-Fuller test:
    print('Results of Dickey-Fuller Test:')
    dftest = adfuller(timeseries, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    print(dfoutput)
```

```python
test_stationarity(ts)
```

```python
fig = plt.figure(figsize=(12,8))
ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(ts, lags=60, ax=ax1)
ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(ts, lags=60, ax=ax2)
```

### First Seasonal Difference

```python
ts_seasonal = ts - ts.shift(52)
ts_seasonal_first_diff = ts_seasonal - ts_seasonal.shift()
```

```python
fig = plt.figure(figsize=(12,8))
ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(ts_seasonal_first_diff[53:], lags=60, ax=ax1)
ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(ts_seasonal_first_diff[53:], lags=60, ax=ax2)
```

```python
test_stationarity(ts_seasonal_first_diff.dropna(inplace=False))
```

### Log First Seasonal Difference

```python
ts_log = np.log(ts)
ts_seasonal_log = ts_log - ts_log.shift(52)
ts_seasonal_first_diff_log = ts_seasonal_log - ts_seasonal_log.shift()
```

```python
fig = plt.figure(figsize=(12,8))
ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(ts_seasonal_first_diff_log[53:], lags=60, ax=ax1)
ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(ts_seasonal_first_diff_log[53:], lags=60, ax=ax2)
```

```python
test_stationarity(ts_seasonal_first_diff_log.dropna(inplace=False))
```

## Make Predictions

```python
prediction_period = 10
train = ts[:-prediction_period]
test = ts[-prediction_period:]
```

### Non-log Predictions

```python
mod = sm.tsa.statespace.SARIMAX(train, trend='n', order=(1,1,0), seasonal_order=(0,1,0,52))
results = mod.fit(disp=0)
```

```python
short_term_predictions = results.forecast(prediction_period)
long_term_predictions = results.forecast(50)
```

```python
print(np.sqrt(mean_squared_error(test, short_term_predictions)))
plt.plot(ts, 'bo')
plt.plot(short_term_predictions, 'r')
ax = plt.subplot()
ax.set_xlim([train.index[-1], long_term_predictions.index[len(short_term_predictions)]])
ax.set_ylim([test[0] - test[0]/10, test[-1]  + test[-1]/10])
plt.show()
```

```python
plt.plot(ts, 'bo')
plt.plot(long_term_predictions, 'r')
plt.show()
```

### Log predictions

```python
mod = sm.tsa.statespace.SARIMAX(np.log(train), trend='n', order=(7,1,0), seasonal_order=(0,1,0,52))
results = mod.fit(disp=0)
```

```python
short_term_log_predictions = np.exp(results.forecast(prediction_period))
long_term_log_predictions = np.exp(results.forecast(50))
```

```python
print(np.sqrt(mean_squared_error(test, short_term_log_predictions)))
plt.plot(ts, 'bo')
plt.plot(short_term_log_predictions, 'r')
ax = plt.subplot()
ax.set_xlim([train.index[-1], long_term_log_predictions.index[len(short_term_log_predictions)]])
ax.set_ylim([test[0] - test[0]/10, test[-1]  + test[-1]/10])
plt.show()
```

```python
plt.plot(ts, 'bo')
plt.plot(long_term_log_predictions, 'r')
plt.show()
```
